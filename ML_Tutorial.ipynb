```python
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
data=load_breast_cancer()
```

```python
print("Target names : ",data['target_names'])
print("First 30 target values : ",data['target'][0:30])
print("Feature names : ",data['feature_names'])
print("Features of first entry : ")
print(data['data'][0])
```

```python
target_names=data['target_names']
target=data['target']
feature_names=data['feature_names']
features=data['data']
```

```python
train_features,test_features,train_target,test_target=train_test_split(features,target,test_size=0.33,
random_state=42) 
```

```python
model=GaussianNB().fit(train_features,train_target)
```

```python
prediction= model.predict(test_features)
print("Actual target values of first 30 test inputs : ",test_target[0:30])
print("Predictions of first 30 target values : ",prediction[0:30])
```

```python
print("Accuracy of predictions : ",accuracy_score(test_target, prediction)*100,"%")
```
